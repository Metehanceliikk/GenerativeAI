```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ithalatDegeri=np.array([20.12, 20.34, 30.01, 30.85, 40.70, 40.55, 40.35, 50.43, 55.60,63.90,63.52])
ihracatDegeri=np.array([30.12, 30.68, 30.71, 37.98, 40.75, 40.58, 51.45, 51.98, 42.60, 53.28,55.88])
yillar=list(range(2014,2025))
```

```python
plt.figure(figsize=(10,7))
plt.subplot(2,1,1)
plt.plot(ithalatDegeri,yillar,color="#f72f2f",marker="o",markersize=8,markerfacecolor="blue")
plt.title("Ülkemizde yıllara göre ithalat değişimi")
plt.xlabel("% Değer")
plt.ylabel("Yıllar")
plt.subplot(2,1,2)
plt.scatter(ihracatDegeri,yillar,color="#921bd3")
plt.title("Ülkemizde yıllara göre ihracat değişimi")
plt.xlabel("% Değer")
plt.ylabel("Yıllar")
plt.tight_layout()
```
