```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
programlamaDilleri=np.array(["Python", "JavaScript","Java","C++", "C#", "Ruby","Swift","Go", "Kotlin","PHP"])
kullananKisiSayisi=np.random.randint(0,2000,10)
```

```python
fig, ax = plt.subplots()
ax.barh(programlamaDilleri,kullananKisiSayisi)
plt.title("2024 yılında en çok kullanılan programlama dilleri")
plt.ylabel("Programlama Dilleri")
plt.xlabel("Kullanan Şirket Sayısı")
plt.show()
```
