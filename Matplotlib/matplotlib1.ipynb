```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sirketDizisi=np.array(["LCW","KOTON","NIKE","ADIDAS","MAVI","DEFACTO","KIGILI","PUMA"])
aylikOrtalamaGelir=np.array([1000,1200,1500,14500,1300,1700,1900,2500])                     #Gelir-Gider verileri girildi
aylikOrtalamaGider=np.random.randint(100,1000,8)       
```

```python
figur,eksenler=plt.subplots(nrows=1,ncols=2,figsize=[10,4])
eksenler[0].plot(sirketDizisi,aylikOrtalamaGelir,label="gelir",color="red",linestyle="-.",marker="o")
eksenler[0].set_xlabel("SIRKETLER")
eksenler[0].set_ylabel("GELIR")
eksenler[0].set_title("Gelir Tablosu")
eksenler[1].plot(sirketDizisi,aylikOrtalamaGider,label="gider",color="blue",linestyle="-.",marker="o")
eksenler[1].set_xlabel("SIRKETLER")
eksenler[1].set_ylabel("GIDER")
eksenler[1].set_title("Gider Tablosu")
plt.tight_layout()
eksenler[0].legend()
eksenler[1].legend()
plt.show()
```
